```python
df = {}
for YEAR in range(1999, 2023):
    df[YEAR] = pd.read_csv(f"https://github.com/nflverse/nflverse-data/releases/download/pbp/play_by_play_{YEAR}.csv.gz")
    print(f"{YEAR} complete.")
```

```python
df[2022].info()
```

```python
drive_df = df[2022] \
    .groupby(["game_id", "drive"]) \
    .agg({'half_seconds_remaining': ['min', 'max']}) \
    .reset_index() \
    .assign(
        drive_within_two_min_warning = lambda d: d[('half_seconds_remaining', 'min')] <= 120
    ) \
    [["game_id", "drive", "drive_within_two_min_warning"]]

drive_df.columns = drive_df.columns.to_flat_index()
drive_df
    
```

```python
drive_df.columns
```

```python
df_tmp = df[2022] \
    .merge(drive_df, left_on=['game_id', 'drive'], right_on=[('game_id', ''), ('drive', '')]) \
    .drop([('game_id', ''), ('drive', '')], axis=1) \
    .rename(
        columns={"('drive_within_two_min_warning', '')": 'drive_within_two_min_warning'}
        )

df_tmp
```
